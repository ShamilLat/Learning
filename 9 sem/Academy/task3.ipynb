```python
import pymorphy2
from razdel import tokenize
import pandas as pd

morph = pymorphy2.MorphAnalyzer()

def normalize_text(text):
    tokens = [token.text for token in tokenize(text)]
    # Лемматизация
    lemmatized_tokens = [morph.parse(token)[0].normal_form for token in tokens if token.isalpha()]
    return ' '.join(lemmatized_tokens)

normalized_sents = []
ids = []
original_texts = []

for doc_id, doc in enumerate(sents):
    for sent_id, sent in enumerate(doc):
        normalized_sents.append(normalize_text(sent))
        original_texts.append(sent)
        ids.append(f'doc{doc_id}_sent{sent_id}')

# Создание DataFrame
data = {
    'id': ids,
    'original_text': original_texts,
    'normalized_text': normalized_sents
}

df = pd.DataFrame(data)

print(df)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
from razdel import tokenize
import pymorphy2

def preprocess(text):
    tokens = [token.text for token in tokenize(text)]
    morph = pymorphy2.MorphAnalyzer()
    return ' '.join([morph.parse(token)[0].normal_form for token in tokens])

documents = updated_df['normalized_text'].values

tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(documents)

def search(query):
    preprocessed_query = preprocess(query)
    query_vector = tfidf_vectorizer.transform([preprocessed_query])

    similarity_scores = cosine_similarity(query_vector, tfidf_matrix)[0]
    ranked_indices = similarity_scores.argsort()[::-1]

    result = {
        "preprocessed_query": preprocessed_query,
        "ranked_sentences": [
            {
                "similarity_score": similarity_scores[idx],
                "original_sentence": updated_df.iloc[idx]["original_text"],
                "preprocessed_sentence": updated_df.iloc[idx]["normalized_text"]
            }
            for idx in ranked_indices
        ]
    }

    return result

```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
from razdel import tokenize
import pymorphy2

def preprocess(text):
    tokens = [token.text for token in tokenize(text)]
    morph = pymorphy2.MorphAnalyzer()
    return ' '.join([morph.parse(token)[0].normal_form for token in tokens])

documents = updated_df['normalized_text'].values

tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(documents)

cntvec_vectorizer = CountVectorizer()
cntvec_matrix = cntvec_vectorizer.fit_transform(documents)

def search_tfidf(query):
    preprocessed_query = preprocess(query)
    query_vector = tfidf_vectorizer.transform([preprocessed_query])

    similarity_scores = cosine_similarity(query_vector, tfidf_matrix)[0]
    ranked_indices = similarity_scores.argsort()[::-1]

    result = {
        "preprocessed_query": preprocessed_query,
        "ranked_sentences": [
            {
                "similarity_score": similarity_scores[idx],
                "original_sentence": updated_df.iloc[idx]["original_text"],
                "preprocessed_sentence": updated_df.iloc[idx]["normalized_text"]
            }
            for idx in ranked_indices
        ]
    }

    return result, similarity_scores

def search_cntvec(query):
    preprocessed_query = preprocess(query)
    query_vector = cntvec_vectorizer.transform([preprocessed_query])

    similarity_scores = cosine_similarity(query_vector, cntvec_matrix)[0]
    ranked_indices = similarity_scores.argsort()[::-1]

    result = {
        "preprocessed_query": preprocessed_query,
        "ranked_sentences": [
            {
                "similarity_score": similarity_scores[idx],
                "original_sentence": updated_df.iloc[idx]["original_text"],
                "preprocessed_sentence": updated_df.iloc[idx]["normalized_text"]
            }
            for idx in ranked_indices
        ]
    }

    return result, similarity_scores
```

```python
import torch
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import pandas as pd

# Получение вектора предложения
def get_sentence_embedding(sentence):
    tokens = tokenizer(sentence, truncation=True, padding=True, return_tensors='pt')
    with torch.no_grad():
        outputs = model(**tokens)
    sentence_embedding = outputs.last_hidden_state.mean(dim=1).numpy()
    return sentence_embedding

def search_model(query):
    query_embedding = get_sentence_embedding(query)
    sentence_embeddings = np.vstack([get_sentence_embedding(sentence) for sentence in updated_df['normalized_text']])

    similarity_scores = cosine_similarity(query_embedding, sentence_embeddings)[0]
    ranked_indices = similarity_scores.argsort()[::-1]

    result = {
        "preprocessed_query": query_embedding,
        "ranked_sentences": [
            {
                "similarity_score": similarity_scores[idx],
                "original_sentence": updated_df.iloc[idx]["original_text"],
                "preprocessed_sentence": updated_df.iloc[idx]["normalized_text"]
            }
            for idx in ranked_indices
        ]
    }

    return result, similarity_scores
```

```python
def ndcg(y_true, y_score, k=10):
    order = np.argsort(y_score)[::-1]
    y_true = np.take(y_true, order[:k])

    dcg = np.sum(y_true / np.log2(np.arange(2, y_true.size + 2)))
    idcg = np.sum(sorted(y_true, reverse=True) / np.log2(np.arange(2, y_true.size + 2)))

    ndcg_value = dcg / idcg
    return ndcg_value
```
